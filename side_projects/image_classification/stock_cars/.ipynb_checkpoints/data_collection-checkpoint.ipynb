# Imports

```python
import os
from pathlib import Path
from natsort import natsorted
import cv2 as cv
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Setting up files and paths

```python
current_working_dir = os.getcwd()
main_project_path = str(Path(current_working_dir).parents[2])
project_data_path = os.path.join(main_project_path, 'side_projects', 'image_classification', 'stock_cars', 'new_data')
project_data_dump_path = os.path.join(main_project_path, 'side_projects', 'data_dumps', 'ImageClassification', 'StockCarShops')
```

```python
paths = {
    'main_project': main_project_path,
    'project_data': project_data_path,
    'project_data_dump': project_data_dump_path
}
```

```python
for k,v in paths.items():
    print(k, v)
```

```python
files = []
for file_name in os.listdir(paths['project_data_dump']):
    files.append(file_name)
    
files = natsorted(files)
files
```

# Segregating data

```python
!ls {paths['project_data']}
```

```python
class_names = ['is_300', 'punto', 'cobalt', 'golf_gti', 'audi_tt', 'audi_a3', 'eclipse', 'audi_a4',
               'supra', 'clio', 'rx8', 'cts', 'mustang_gt', 'evolution', 'sl500', 'gto', 
               'monaro', 'cayman', 'wrx', 'rx7', 'clk500', 'elise', 'db9', 'carrera', 
               'viper', 'gallardo', 'turbo', 'corvette', 'murcielago', 'ford_gt', 'mercedes_benz', 'carrera_gt']
class_names
```

```python
for class_ in class_names:
    class_dir = os.path.join(paths['project_data'], class_)
    train_dir = os.path.join(class_dir, 'train')
    test_dir = os.path.join(class_dir, 'test')
    if not os.path.exists(class_dir):
        !mkdir {class_dir}
    if not os.path.exists(train_dir):
        !mkdir {train_dir}
    if not os.path.exists(test_dir):
        !mkdir {test_dir}
    
!ls {paths['project_data']}
```

```python
for dirpath, dirnames, filenames in os.walk(paths['project_data']):
    print('There are {} directories, {} files in {}'.format(len(dirnames), len(filenames), dirpath))
```

```python
paths
```

```python
for i in range(1, 801, 25):
    block = range(i, i+25)
    for file_number in block:
        class_train_path = os.path.join(paths['project_data'], class_names[i//25], 'train')
        class_test_path = os.path.join(paths['project_data'], class_names[i//25], 'test')
        
        test_images = random.sample(block, 5)
        train_images = list(set(block) - set(test_images))
#         print(test_images, train_images)
        
        for test_image in test_images:
            src_path = os.path.join(paths['project_data_dump'], str(test_image)+'.jpg')
            dst_path = os.path.join(class_test_path, str(test_image)+'.jpg')
            !cp {src_path} {dst_path}
            print(src_path, dst_path, sep='\n')
            
        for train_image in train_images:
            src_path = os.path.join(paths['project_data_dump'], str(train_image)+'.jpg')
            dst_path = os.path.join(class_train_path, str(train_image)+'.jpg')
            !cp {src_path} {dst_path}
            print(src_path, dst_path, sep='\n')
            
        break
    
```

```python
for dirpath, dirnames, filenames in os.walk(paths['project_data']):
    print('There are {} directories, {} files in {}'.format(len(dirnames), len(filenames), dirpath))
```

```python
for dirpath, dirnames, filenames in os.walk(paths['project_data']):
    if len(filenames) > 0:
        for filename in filenames:
            file_path = os.path.join(dirpath, filename)
            image = mpimg.imread(file_path)
            image = cv.cvtColor(image, cv.COLOR_RGBA2RGB)
            image = image[175:525, 325:950]
#             plt.imshow(image)
            plt.imsave(file_path, image)
            print(type(image), image.shape)
        
```
